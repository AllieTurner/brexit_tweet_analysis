# Is there a shift in themes throughout the day?

The easiest way to find out the themes of the discussion is to look at the hashtags. Let's look at the top ten Hashtag for each hour of the 23rd of September 2017. To not consider duplicates, the Hashtags are slightly adjusted.

```python
def fix_hashtag(hashtag):
    """Because of capitliation and similar errors some hashtags are the same and need
       to be adjusted (eg. 'EU' and 'Eu' and 'eu' should all be the same)"""
    if (hashtag.upper() == "#EU"):
        return "#EU"
    if (hashtag.upper() == "#FLORENCESPEECH"):
        return "FlorenceSpeech"
    fixed = hashtag[0] + hashtag[1].upper() + hashtag[2:]
    return fixed


def get_used_hashtags_from_entities(current_data):
    """There is another way of getting the used Hashtags through the entities dictionary, without
    looking at the whole text, but entities is a horrible array of dictionaries"""
    hashtags = []
    #print(current_data['entities']['hashtags'])
    for i in current_data['entities']:
        for hashtag in i['hashtags']:
            hashtags.append(hashtag['text'])
    hashtag_occurences = {}
    for hashtag in hashtags:
        hashtag = fix_hashtag("#" + hashtag)
        if hashtag in hashtag_occurences.keys():
            hashtag_occurences[hashtag] +=1
        else:
            hashtag_occurences[hashtag] = 1
    return pd.Series(hashtag_occurences)
        
def plot_most_used_hashtags(hashtag_series):
    """Plots the top ten hashtags of a Data Series"""
    #Plot the top ten used hashtags
    hashtag_series.sort_values(inplace=True,ascending=False)
    ax = sns.barplot(x=hashtag_series[:10].index,y=hashtag_series[:10])
    for tick in ax.get_xticklabels():
        tick.set_rotation(90)
    sns.plt.show()

from glob import glob
def get_top_ten_of_day(day):
    """Given a day (which should be a folder containing the hourly json files),
    computes the top ten hashtags of each day and returns a list of each hashtag and its position on each day"""
    positions = []
    no_files = 0
    all_hashtags = set()
    for f in glob(day+"/*.json"):
        positions.append({})
        hashtag_series = get_used_hashtags_from_entities(pd.read_json(f,lines=True))
        #Get the ten most used Hashtags and mark their position
        top_ten = hashtag_series.sort_values(ascending=False)[:10]
        for hashtag in top_ten.index:
            all_hashtags.add(hashtag)
        for i in range(0,10):
            if top_ten.index[i] in positions[no_files].keys():
                positions[no_files][top_ten.index[i]].append([i + 1])
            else:
                positions[no_files][top_ten.index[i]] = [i + 1]
        no_files += 1

    #get Positions for each Hashtag
    position_of_hashtags = {}
    for hashtag in all_hashtags:
        position_of_hashtags[hashtag] = []
        for top_ten in positions:
            if hashtag in top_ten.keys():
                position_of_hashtags[hashtag].append(top_ten[hashtag][0])
            else:
                position_of_hashtags[hashtag].append(np.nan)
    return position_of_hashtags

def categorize_hashtag(hashtag):
    return {
        '#UKIP' : 'red',
        '#Leave': 'red',
        '#Asean': 'grey',
        '#UK'   : 'green',
        '#Iran' : 'grey',
        '#Home' : 'red',
        "#OpKillingBay" : 'grey',
        "#ContentWriting" : 'grey',
        "#FlorenceSpeech" : 'green',
        "#Un" : 'blue',
        "#Brexit" : "green",
        "#Europe" : "green",
        "#Shipping" : "grey",
        "#Decoration" : "grey",
        "#Nato" : "green",
        "#Government" : "green",
        "#Remain" : "blue",
        "#ToriesOut" : "blue",
        "#Us" : "grey",
        "#Referendum": "green",
        "#News" : "green",
        "#CorporateManslaughter": "grey",
        "#MayMustGo" : "blue",
        "#Nz" : "grey",
        "#WeWe" : "green",
        "#Occitània" : "grey",
        "#Multilingualseo" : "grey",
        "#RefugeesWelcome" : "blue",
        "#EuropeanUnion" : "green",
        "#Fair" : "grey",
        "#SSC" : "grey",
        "#ContentMarketing" : "grey",
        "#Textile" : "grey",
        "#ConflictofInterest" : "blue",
        "#JC4PM…" : "green",
        "#Sale" : "grey",
        '#Brexit?"' : "blue",
        "#Obamacare" : "grey",
        "#Multilingual" : "grey",
        "#Mayhem" : "blue",
        "#TheresaMay" : "green",
        "#EU" : "green",
        "#Catalunya" : "grey",
        "#StopBrexit" : "blue",
        "#SEO" : "grey",
        "#EuropeUnion" : "green",
        "#CatNord" : "grey",
        "#NHS" : "green",
        "FlorenceSpeech" : "green",
        "#FaroeIslands" : "grey",
        "#ToryElectionFraud" : "blue",
        "#Conservatives" : "red",
        "#LabourIn" : "green",
        "#WalkAwayMay" : "blue",
        "#Au" : "grey",
        "#BankPO" : "grey",
        "#Blog" : "grey",
        "#USA" : "grey",
        "#Catalonia" : "grey"
        
    }[hashtag]

from matplotlib.lines import Line2D
from itertools import cycle
def get_markers_and_colors(position_of_hashtags,categorized=False,palette="Set2"):
    """Given a set of hashtags, the function returns a colormap and a markermap from the hashtags"""
    colorset = dict()
    hashtags = position_of_hashtags.keys()
    if categorized:
        for hashtag in hashtags:
            colorset[hashtag] = categorize_hashtag(hashtag)
    else:
        colorset = dict(zip(hashtags,sns.color_palette("Set2", len(hashtags))))
    markers = []
    for m in Line2D.markers:
        if(type(m) == type("")):
            markers.append(m)
    markerset = dict(zip(hashtags,cycle(markers)))
    return colorset,markerset
def plot_top_ten_hashtag_of_day(position_of_hashtags,categorized=False):
    colorset,markerset = get_markers_and_colors(position_of_hashtags,categorized=categorized)
    #Plot each Hashtag as a line of positions
    fig, ax = sns.plt.subplots(figsize=(37,5),dpi=250)
    for hashtag,positions in position_of_hashtags.items():
        file_no = []
        for i in range(1,len(positions)):
            file_no.append(i)
        #print(positions)
        y = pd.Series(file_no)
        ax = sns.pointplot(y=positions,x=y,color=colorset[hashtag],label=hashtag,markers=markerset[hashtag])
        ax.set_ylim(11,0)
    leg = ax.legend(handles=ax.lines[::len(positions)+1], labels=position_of_hashtags.keys(),
              bbox_to_anchor=(1.0, 1.05),ncol=2, fancybox=True, shadow=True,numpoints = 1 )
    for handle, text in zip(leg.legendHandles, leg.get_texts()):
        handle.set_marker(markerset[text.get_text()])
        handle.set_color(colorset[text.get_text()])
    ax.set_title("Top 10 Hashtags in the first 24 Hours of the 23rd of September")
    ax.set(xlabel="Hour of the day (23rd of September)",ylabel="Position in the Top Ten of Hashtags used")
    sns.plt.show()
```

```python
#Get all the Hashtag-Data from 24 hours on the 23rd of September
position_of_hashtags = get_top_ten_of_day("2017-09-23")
#Plot the Hashtags with different colors and markers
plot_top_ten_hashtag_of_day(position_of_hashtags)
```

During the Whole day, the #Brexit and #EU Hashtags stay on top, with the themes rapidly changing, especially in the lower ranks of the popular hashtags. Let's now categorized these into Pro-Brexit, Anti-Brexit, Neutral and unrelated

```python
#Do the same plot, but make red pro brexit, blue anti brexit, green neutral and grey for unrelated
plot_top_ten_hashtag_of_day(position_of_hashtags,True)
```

While neutral Hashtags are dominating the ranks, as they are used in combination with both positive and negative hashtags. The Anti Brexit is a bit more dominant, especially during the later hours (15:00 and onward). At 14:00 there was a very shortlived hashtag-campaign from the anti brexit front which still was able to influence the later hours.

# Where are Users tweeting from

We can use two different metrics to figure out where users are from. Timezones and location. As only a really
small fraction of tweets are geotagged, we rely on self set locations. Timezones however, are set by twitter.

```python
location_data = set()
location_coords = {}
```

Firstly, location data is feed through the gmaps api to produce locations from strings (geocodes). These are
saved, so you don't need to load them everytime.

```python
#As you can't see the map, this and the next cell needs tu be rerun
import geopy,gmaps
from geopy.geocoders import Nominatim
import os
if not os.path.exists("locations.csv"):
    geolocator = Nominatim()
    locations = []
    for i in data['user']:
        location = i['location']
        geocode = None
        if location not in location_data:
            try :
                geocode = geolocator.geocode(location)
                print("Location found ",(geocode.latitude, geocode.longitude),"for",location)
                location_data.add(location)
                location_coords[location] = (geocode.latitude, geocode.longitude)
                locations.append((geocode.latitude, geocode.longitude))
            except Exception as e:
                print("Location not found, or timed out for",location, "Error:",e)
        else:
            geocode = location_coords[location]
            locations.append((geocode[0],geocode[1]))
    location_occurences_series = pd.Series(location_coords)
    location_occurences_Series.to_csv("locations.csv")
else:
    location_coords = pd.read_csv("locations.csv")
```

# Reflection

The Data is really complex and contains so many nested dictionaries as columns. This makes the direct work on the data frame much harder. Optimally some columns should be added to the whole dataframe, and many others removed. As all the tweets where recorded as they came in, we don't know the reception (favourite count, quote count, replies) of each tweet without iterating over every other tweet (which wouldn't be viable). I do like the richness of the data, especially the user column. This will make it easy to try to figure out who are the people in the center of discussion shaping the discourse. The amount of the data is also very interesting. I chose on of the lowest-frequented hours and still got over 2k tweets. The most frequented hours contain 100k tweets. I tried to write my code in a way that it can be reused for all the days/hours, but the amount of columns actually changes depending on the day, which makes it much harder to write consistent code on the dataset. This is probably because either the twitter api changed between June of last year and september of this year or the people capturing the data chose to capture more columns.
